# <b> Model Development

```python
lr=LinearRegression()
x=df[['Review Rating']]
y=df['Purchase Amount (USD)']
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
df_model=df.copy()

df_model=df_model.dropna(subset=['Purchase Amount (USD)', 'Review Rating', 'Payment Method', 'Item Purchased', 'Date Purchase']) 
df_model['Purchase Month']=df_model['Date Purchase'].dt.month
df_model['Purchase_day_of_week']=df_model['Date Purchase'].dt.dayofweek

categorical_columns=['Payment Method', 'Item Purchased']
df_model_dumy=pd.get_dummies(df_model[['Payment Method', 'Item Purchased']],drop_first=True)
df_model_dumy=df_model_dumy.astype(int)
df_model=pd.concat([df_model,df_model_dumy],axis=1)


feature_cols = ['Review Rating', 'Purchase Month', 'Purchase_day_of_week'] + \
               [col for col in df_model.columns if col.startswith('Payment Method_') or col.startswith('Item Purchased_')] 

x=df_model[feature_cols]
y=df_model['Purchase Amount (USD)']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) 
lr=LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_test)

print(f'R2 Score for predicting Purchase Amount (USD):{lr.score(x_test,y_test)}')
```
